```python
import pandas as pd
from matplotlib import pyplot as plt

def lexical_diversity(text):
    return len(set(text)) / len(text)

df = pd.DataFrame(text1.vocab().most_common(50), columns=['word', 'frequency'])

# Plot cumulative frequency graph
df['cumulative_frequency'] = df['frequency'].cumsum()
df['cumulative_percentage'] = df['cumulative_frequency'] / len(text1) * 100

plt.plot(df['cumulative_percentage'])
plt.xlabel('Word rank')
plt.ylabel('Cumulative percentage')
plt.show()
```
